## Task 1: Source the Data Set - Stated Aim

The Iris dataset is one of the most famous datasets in machine learning, it was compiled by the British biologist/statistician Ronald Fisher in 1936. It contains 150 iris flower samples from 3 different species & 4 features per sample.

**Species of iris flower:**
 - Setosa
 - Versicolor
 - Virginica

**Features of iris flower:**
- Sepal Length
- Sepal Width
- Petal Length
- Petal Width


To start, we import the Iris dataset from the Scikit-Learn library into the Jupyter notebook.

This function loads the raw csv file from Scklit-Learn library into the Jupyter notebook. 
The iris_dataset is stored as a bunch object and it contains several keys that we can call in the notebook, including:

**'data'**: The feature data (petal/sepal measurements)

```python
## Task 8: Compute Correlations - Code Cell


#Applicable columns are first 4, dataset will include the added columns of species', 'species_name' so subset needed.
#iloc used similar to above for aplicabale columsn only
df_subset=df.iloc[:, 0:4]
#print(df_subset)


# Correlation Coefficient . Numeric Only so 
corr_matrix = df_subset.corr()         # Reference : https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.corr.html

print(corr_matrix)


# heatmap using matplotlib

# Create figure, axis.
fig, ax = plt.subplots()

# Plot heatmap
image = ax.imshow(corr_matrix,cmap='coolwarm')

# Add colorbar
fig.colorbar(image, ax=ax, label='Correlation Coefficient')

# Labels

ax.set_xticks(np.arange(len(corr_matrix.columns)))
ax.set_yticks(np.arange(len(corr_matrix.index)))
ax.set_xticklabels(corr_matrix.columns, rotation=45, ha="right")
ax.set_yticklabels(corr_matrix.index)

#Add title
plt.title('Correlation Heatmap')

# Plot
plt.tight_layout()
plt.show()
```

## Task 8: Compute Correlations - Analysis

Correlation matrix & heatmap is used to show the strength of relationship between each feature. Range of values go from -1 to 1.

 - +1 - Full positive correlation - as one feature increases so will the other
 - 0 - No correlation - no relationship, features are independent
 - Full negative correlation - as one feature decreases so will the other

The correlation heatmap is a visual representation of the correlation matrix. Colours, on a sliding scale of dark red (strong relationship) & dark blue (weak relationship) are used to depcit the strength of the correlation between each of the 4 features

 **Sepal Length vs other features**
 - Strong Positive correlation with petal length (0.87) & petal width (0.81). Larger the sepal length the more 
 - Negative correlation with sepal width (-0.11)- the larger the sepal length the smaller the sepal width is likely to be.

**Sepal Width vs other features**
- Negative correlation with petal length(-0.43) , petal width(-0.37) & Sepal Length(-0.12) . As sepal width increases the other features would decrease. It should be noted that the negative correlation is very weak with sepal length(-0.12)

**Petal Length vs other features**
- Strong positive relationshsip with petal width (0.96) & sepal length(0.87). As petal length increases so would petal width & sepal length nearly proportionately.
- Negative relationship with sepal width (-0.43). As petal length increases , we would expect sepal width to get smaller

**Petal Width vs other features**
- Strong positive relationship with petal length (0.96) & sepal length(0.82). As petal width increases so would petal length & sepal length nearly proportionately.
- Negative relationship with sepal width (-0.37). As petal width increases , we would expect sepal width to get smaller

Overall, we can see that:

Sepal Length is strongly correlated with petal length & width  
Petal width is strongly correlated with sepal length & petal length  
Petal length is strongly correlated with sepal length & petal width  
Sepal width has a negative correlation with petal length/width & sepal length
